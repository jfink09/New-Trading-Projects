```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
NEX_funds = [282,289,294,323,327,372,403,400]
AMR_funds = [66,75,73,82,133,166,268,303]
STNG_funds = [208,224,229,240,231,211,244,265]
PBF_funds = [329,328,364,353,368,413,484,525]
TH_funds = [93,85,99,118,129,130,149,170]
AR_funds = [290,380,430,478,524,593,694,716]
TNK_funds = [95,95,156,175,154,176,210,203]
OXY_funds = [1139,1204,1241,1294,1389,1548,1700,1634]
CRK_funds = [136,152,206,234,274,305,336,344]
TDW_funds = [161,159,166,187,169,195,234,252]

a = NEX_pct_chg = np.diff(NEX_funds) / NEX_funds[:-1] * 100
b = AMR_pct_chg = np.diff(AMR_funds) / AMR_funds[:-1] * 100
c = STNG_pct_chg = np.diff(STNG_funds) / STNG_funds[:-1] * 100
d = PBF_pct_chg = np.diff(PBF_funds) / PBF_funds[:-1] * 100
e = TH_pct_chg = np.diff(TH_funds) / TH_funds[:-1] * 100
f = AR_pct_chg = np.diff(AR_funds) / AR_funds[:-1] * 100
g = TNK_pct_chg = np.diff(TNK_funds) / TNK_funds[:-1] * 100
h = OXY_pct_chg = np.diff(OXY_funds) / OXY_funds[:-1] * 100
i = CRK_pct_chg = np.diff(CRK_funds) / CRK_funds[:-1] * 100
j = TDW_pct_chg = np.diff(TDW_funds) / TDW_funds[:-1] * 100
```

```python
#df.reset_index
df = pd.DataFrame(list(zip(a,b,c,d,e,f,g,h,i,j))) 
df.columns = ["NEX","AMR","STNG","PBF","TH","AR",'TNK','OXY','CRK','TDW']
df
```
