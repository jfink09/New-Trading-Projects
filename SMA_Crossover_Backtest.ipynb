```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import plotly.express as px
```

```python
SPY = yf.download("SPY",period="2y",progress=False)
SPY
```

```python
fig = px.line(SPY, y="Adj Close", title='SPY Price', labels = {'Adj Close':'SPY Close Price(in USD)'})
fig.update_layout({
'plot_bgcolor': '#FFFFFF',
'paper_bgcolor': '#FFFFFF',
})
fig.show()
```

```python
window1 = 10
sma1 = pd.DataFrame()
sma1['Adj Close'] = SPY['Adj Close'].rolling(window = window1).mean()
sma1
```

```python
window2 = 21
sma2 = pd.DataFrame()
sma2['Adj Close'] = SPY['Adj Close'].rolling(window = window2).mean()
sma2
```

```python
fig.add_scatter(x=sma1.index,y=sma1['Adj Close'], mode='lines',name='SMA'+str(window1))
fig.add_scatter(x=sma2.index,y=sma2['Adj Close'], mode='lines',name='SMA'+str(window2))
fig.show()
```

```python
data = pd.DataFrame()
data['SPY'] = SPY['Adj Close']
data['SMA'+str(window1)] = sma1['Adj Close']
data['SMA'+str(window2)] = sma2['Adj Close']
data
```

```python
def dualMACrossover(data):
    sigPriceBuy = []
    sigPriceSell = []
    flag = -1 # Flag denoting when the 2 moving averages crossed each other
    for i in range(len(data)):
        if data['SMA'+str(window1)][i] > data['SMA'+str(window2)][i]:
            if flag != 1:
                sigPriceBuy.append(data['SPY'][i])
                sigPriceSell.append(np.nan)
                flag = 1
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        elif data['SMA'+str(window1)][i] < data['SMA'+str(window2)][i]:
            if flag!=0:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(data['SPY'][i])
                flag=0
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        else:
            sigPriceBuy.append(np.nan)
            sigPriceSell.append(np.nan)
    return (sigPriceBuy,sigPriceSell)
```

```python
buy_sell = dualMACrossover(data)
data['BuySignalPrice'] = buy_sell[0]
data['SellSignalPrice'] = buy_sell[1]
```

```python
import plotly.graph_objects as go

fig = px.line(data, y="SPY", title='Strategy Visualization', labels = {'index':'Date'})
fig.update_layout({
'plot_bgcolor': '#FFFFFF',
'paper_bgcolor': '#FFFFFF',
})
fig.add_scatter(x=data.index,y=data['SMA'+str(window1)], mode='lines',name='SMA'+str(window1))
fig.add_scatter(x=data.index,y=data['SMA'+str(window2)], mode='lines',name='SMA'+str(window2))

fig.add_trace(go.Scatter(mode="markers", x=data.index, y=data.BuySignalPrice, marker_symbol='triangle-up',
                           marker_line_color="#00FF00", marker_color="#00FF00", 
                           marker_line_width=1, marker_size=8, name='Buy'))

fig.add_trace(go.Scatter(mode="markers", x=data.index, y=data.SellSignalPrice, marker_symbol='triangle-down',
                           marker_line_color="#FF0000", marker_color="#FF0000", 
                           marker_line_width=1, marker_size=8, name='Sell'))
fig.show()
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
```

```python
class DualMACrossover(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, window1)
        self.ma2 = self.I(SMA, price, window2)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()


bt = Backtest(SPY, DualMACrossover,
              exclusive_orders=True)
stats = bt.run()
bt.plot()
```
