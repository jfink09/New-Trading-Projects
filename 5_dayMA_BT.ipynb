```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.express as px
from scipy.optimize import curve_fit
```

```python
SPY = yf.download("SPY",start="2010-01-01",end=None,progress=False)
SPY
```

```python
# Plot the stock (SPY) chart
fig = px.line(SPY, y="Adj Close", title='SPY Price', labels = {'Adj Close':'SPY Close Price(in USD)'})
fig.update_layout({
'plot_bgcolor': '#FFFFFF',
'paper_bgcolor': '#FFFFFF',
})
fig.show()
```

```python
DAYS_5 = 5
sma_5 = pd.DataFrame()
sma_5['Adj Close'] = SPY['Adj Close'].rolling(window = DAYS_5).mean()
sma_5
```

```python
DAYS_1 = 1
sma_1 = pd.DataFrame()
sma_1['Adj Close'] = SPY['Adj Close'].rolling(window = DAYS_1).mean()
sma_1
```

```python
fig.add_scatter(x=sma_5.index,y=sma_5['Adj Close'], mode='lines',name=str(DAYS_5)+'SMA')
fig.add_scatter(x=sma_1.index,y=sma_1['Adj Close'], mode='lines',name=str(DAYS_1)+'SMA')
fig.show()
```

```python
from numpy.core.fromnumeric import size
data = pd.DataFrame()
data['SPY Close'] = SPY['Adj Close']
data[str(DAYS_5)+'SMA'] = sma_5['Adj Close']
data[str(DAYS_1)+'SMA'] = sma_1['Adj Close']
data
```

```python
def dualMACrossover(data):
    sigPriceBuy = []
    sigPriceSell = []
    flag = -1 # Flag denoting when the 2 moving averages crossed each other
    for i in range(len(data)):
        if data[str(DAYS_5)+'SMA'][i] > data[str(DAYS_1)+'SMA'][i]:
            if flag != 1:
                sigPriceBuy.append(data['SPY Close'][i])
                sigPriceSell.append(np.nan)
                flag = 1
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        elif data[str(DAYS_5)+'SMA'][i] < data[str(DAYS_1)+'SMA'][i]:
            if flag!=0:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(data['SPY Close'][i])
                flag=0
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        else:
            sigPriceBuy.append(np.nan)
            sigPriceSell.append(np.nan)
    return (sigPriceBuy,sigPriceSell)
```

```python
buy_sell = dualMACrossover(data)
data['BuySignalPrice'] = buy_sell[0]
data['SellSignalPrice'] = buy_sell[1]
```

```python
import plotly.graph_objects as go

fig = px.line(data, y="SPY Close", title='Strategy Visualization', labels = {'index':'Date'})
fig.update_layout({
'plot_bgcolor': '#FFFFFF',
'paper_bgcolor': '#FFFFFF',
})
fig.add_scatter(x=data.index,y=data[str(DAYS_5)+'SMA'], mode='lines',name=str(DAYS_5)+'SMA')
fig.add_scatter(x=data.index,y=data[str(DAYS_1)+'SMA'], mode='lines',name=str(DAYS_1)+'SMA')

fig.add_trace(go.Scatter(mode="markers", x=data.index, y=data.BuySignalPrice, marker_symbol='triangle-up',
                           marker_line_color="#00FF00", marker_color="#00FF00",
                           marker_line_width=1, marker_size=8, name='Buy'))

fig.add_trace(go.Scatter(mode="markers", x=data.index, y=data.SellSignalPrice, marker_symbol='triangle-down',
                           marker_line_color="#FF0000", marker_color="#FF0000",
                           marker_line_width=1, marker_size=8, name='Sell'))
fig.show()
```

```python
# Import backtesting libraries
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
```

```python
class DualMACrossover(Strategy):
    def init(self):
        price = self.data.Close
        self.ma5 = self.I(SMA, price, DAYS_5)
        self.ma1 = self.I(SMA, price, DAYS_1)

    def next(self):
        if crossover(self.ma5, self.ma1):
            self.buy()
        elif crossover(self.ma1, self.ma5):
            self.sell()


bt = Backtest(SPY, DualMACrossover,
              exclusive_orders=True)
stats = bt.run()
bt.plot()
```
